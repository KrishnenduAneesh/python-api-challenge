### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
%%capture --no-display
import folium

# Create a map centered around the first city in the DataFrame
map_center = [city_data_df['Lat'].iloc[0], city_data_df['Lng'].iloc[0]]
mymap = folium.Map(location=map_center, zoom_start=4)

# Add a marker for each city with the size of the point representing humidity
for index, row in city_data_df.iterrows():
    folium.CircleMarker(
        location=[row['Lat'], row['Lng']],
        radius=row['Humidity'] / 2,  # Adjust the scale as needed
        popup=f"{row['City']}, {row['Country']}",
        #color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.2
    ).add_to(mymap)

# Save the map to an HTML file
mymap.save("C:/Users/krish/OneDrive/Desktop/Class_Activities/Challenge/Challenge_6/Starter_Code/output_data/city_map.html")

# Display the map
mymap
```

```python
from IPython.display import display
import folium

# Set the initial coordinates for the map (you can adjust this based on your data)
map_center = [0, 0]
my_map = folium.Map(location=map_center, zoom_start=2)

# Add markers for each city in the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    hotel_name = row["Hotel Name"]
    city_name = row["City"]
    country_code = row["Country"]
    lat = row["Lat"]
    lng = row["Lng"]

    # Create a popup message for the marker
    popup_msg = f"{hotel_name}, {city_name}, {country_code}"

    # Add a marker for each city
    folium.Marker(
        location=[lat, lng],
        popup=popup_msg,
        icon=folium.Icon(color="blue")
    ).add_to(my_map)

# Display the map in the Jupyter Notebook
display(my_map)

```
